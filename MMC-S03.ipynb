Equação (1):

δ(d)=23.45°⋅sin(365/360°⋅(d−81))

onde d é o número do dia do ano

Equação (2):

H(d)=cos(ϕ)⋅cos(δ(d))⋅cos(h)+sin(ϕ)⋅sin(δ(d))

Como o enunciado pede considerar h=0h = 0h=0 (meio-dia local) e I0=1I_0 = 1I0=1, fica simplificado para calcular a irradiação relativa em cada dia.


```python
import numpy as np
import matplotlib.pyplot as plt

latitude = np.radians(-23.5)

dias = np.arange(1, 366)

delta = np.radians(23.45) * np.sin(np.radians((360/365) * (dias - 81)))

H = np.cos(latitude) * np.cos(delta) + np.sin(latitude) * np.sin(delta)

plt.figure(figsize=(10,5))
plt.plot(dias, H, label="São Paulo (Lat -23.5°)")
plt.xlabel("Dia do ano")
plt.ylabel("Irradiação relativa (unid. adimensional)")
plt.title("Energia Solar Teórica ao Meio-Dia")
plt.legend()
plt.grid(True)
plt.show()
```

Atividade 2 (6 pontos) - Explorando a Energia Solar com Dados da NASA

Nesta parte, vamos compreender como a energia solar varia com latitude, hora do dia e época do ano. Também vamos Simular a produção de um sistema fotovoltaico em uma cidade.

Para isso, vamos revisar os conceitos de latitude e longitude.

Latitude é o ângulo formado entre o Equador e um ponto estimado. Todos os pontos do Equador possuem latitude geográfica igual a 0º. Pontos situados ao norte do equador têm latitudes maiores que 0º, variando até 90º, que é a latitude do polo geográfico norte. Da mesma forma variam as latitudes ao sul do equador terrestre, desde 0º a 90º, latitude do polo geográfico sul. Para se diferenciar os valores, atribui-se sinal positivo para as latitudes norte e negativo para as latitudes sul

```python
import numpy as np
import matplotlib.pyplot as plt

dados = {
    'JAN': 5.2289, 'FEB': 5.4396, 'MAR': 4.9123, 'APR': 4.3562,
    'MAY': 3.5225, 'JUN': 3.2578, 'JUL': 3.4793, 'AUG': 4.1513,
    'SEP': 4.4232, 'OCT': 4.9058, 'NOV': 5.2334, 'DEC': 5.4898
}

meses = list(dados.keys())
rad_dia = np.array(list(dados.values()))

dias_mes = [31,28,31,30,31,30,31,31,30,31,30,31]
rad_mes = rad_dia * dias_mes

area = 15     # m²
eficiencia = 0.14
prod_mes = rad_mes * area * eficiencia

plt.figure(figsize=(10,5))
plt.bar(meses, prod_mes, color='orange')
plt.xlabel("Mês")
plt.ylabel("Energia Gerada (kWh/mês)")
plt.title("Produção Mensal do Sistema Fotovoltaico (15 m², 14% eficiência)")
plt.show()

prod_anual = prod_mes.sum()
prod_media = prod_mes.mean()

print("Produção Anual:", round(prod_anual,2), "kWh")
print("Produção Média Mensal:", round(prod_media,2), "kWh")
```
